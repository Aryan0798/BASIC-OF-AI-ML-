```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=df)
plt.savefig('scatterplot.png')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=df)
plt.savefig('Hueparameter')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=df)
plt.savefig('Styleplot')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data=df)
plt.savefig('Sizeparameterplot')
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=df)
plt.savefig('ScatterPlot1')
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=df)
plt.savefig('Strip.png')
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=df)
plt.savefig('swarm.png')
```

```python
#Boxplot
sns.boxplot(df['tip'])
plt.savefig('SimpleBoxplot')
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=df)
plt.savefig('MultiBoxplot')
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=df)
plt.savefig('MultiBoxplot with Hue')
```

```python
sns.violinplot(df['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=df)
plt.savefig('ViolinPlot.png')
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=df)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',estimator=np.median,data=df)
plt.savefig('Catplot.png')
```

```python
sns.catplot(x='sex',kind='count',hue='smoker',data=df)
plt.savefig('CountPlot.png')
```

```python
#DistPlots
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
sns.distplot(titanic['age'])
plt.savefig('Distplot.png')
```

```python
sns.distplot(titanic['age'],bins=10,kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
plt.savefig('Distplot:KDE,HIST,RUG')
```
