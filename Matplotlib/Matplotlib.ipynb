```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
arr1=[25,26,27,28,29]
arr2=[100000,150000,350000,450000,750000]
arr3=[150000,250000,450000,550000,950000]
```

```python
plt.plot(arr1,arr2,label='ML')
plt.plot(arr1,arr3,label='AI')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary with Age')
plt.legend()
plt.savefig('plot.png')
```

```python
slice=['60','40']
labels=['sixty','forty']
plt.title('Pie Chart')
plt.pie(slice,labels=labels,autopct='%1.1f%%')
plt.savefig('pie.png')
```

```python
mini=[1,2,3,4,5,6,7,8,9]
p1=[1,1,1,2,3,3,4,4,5]
p2=[1,2,2,2,2,3,4,4,5]
p3=[1,1,2,3,3,4,5,5,5]
labels=['player1','player2','player3']
```

```python
plt.stackplot(mini,p1,p2,p3,labels=labels)
plt.legend()
plt.savefig('stackplot.png')
```

```python
plt.plot(arr1,arr2,label='ML')
plt.plot(arr1,arr3,label='AI')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary with Age')
overall_median=350000
plt.fill_between(arr1,arr2,overall_median,interpolate=True,alpha=0.5)
plt.legend()
plt.savefig('Advanceplot.png')
```

```python
ages=[18,19,21,23,26,29,33,35,36,40,45]
```

```python
plt.hist(ages,bins=5)
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.tight_layout()
plt.show()
plt.savefig('Histogram.png')
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(p1,p2)
cbar=plt.colorbar()
cbar.set_label('satisfaction')
plt.savefig('Scatterplot.png')
```

```python
import matplotlib.dates as mdl
import matplotlib.pyplot as plt
from datetime import datetime,timedelta
```

```python
dates=[
    datetime(2025,5,20),
    datetime(2025,5,21),
    datetime(2025,5,22),
    datetime(2025,5,23),
    datetime(2025,5,24),
    datetime(2025,5,25),
    datetime(2025,5,26),
]
y=[0,1,2,3,4,5,6]
plt.plot_date(dates,y)
plt.gcf().autofmt_xdate()
date_format=mdl.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.savefig('Date.png')
```
